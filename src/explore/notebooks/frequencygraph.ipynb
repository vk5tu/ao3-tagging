# Frequency graphs of variables

Intended for publication.

The Top30-minus graphs need post-production work in Inkscape to set the colour of the font of the unprinted rows to gray rather than black.

## Load the data into dataframe `df`

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import ds9
df = ds9.df()
```

## relationshipspairslash

```python
relationshipspairslash_df = ds9.explode(df[['relationshipspairslash']], 'relationshipspairslash')
relationshipspairslash_type = ds9.strtotype(relationshipspairslash_df,'relationshipspairslash')
relationshipspairslash_df['relationshipspairslash'] = relationshipspairslash_df['relationshipspairslash'].astype(relationshipspairslash_type)
relationshipspairslash_df = relationshipspairslash_df.drop('id', axis=1)
```

```python
### Frequency graph of relationships
sns.set(rc = {'figure.figsize':(20,5)})
fig = sns.countplot(data=relationshipspairslash_df, x='relationshipspairslash')
fig.set(title='Number of works in which a slash-pairing appeared')
fig.set(xlabel='Slash-pair of characters')
fig.set(xticklabels=[])
fig.tick_params(bottom=False)
fig.set(ylabel='Works')
fig.figure.savefig('frequencygraph-relationshipspairslash-all.svg')
fig
```

## Top 30 relationshippairslash

```python
r_df = ds9.keepcolumns(df, 'relationshipspairslash')
r_df = ds9.explode(r_df, 'relationshipspairslash')
r_df = ds9.keepcolumns(r_df, 'relationshipspairslash')
                       
# Remove non-cast/non-cast
index = r_df[r_df['relationshipspairslash'] == 'non-cast/non-cast'].index
r_df = r_df.drop(index)
                       
# Find top 30
freq_df = r_df['relationshipspairslash'].value_counts()
freq_list = list(freq_df.index)
top30_list = freq_list[0:30]

# Keep only top 30 in dataframe
keep_df = pd.DataFrame({'relationshipspairslash': top30_list}, dtype = relationshipspairslash_type)
top30_df = r_df[r_df.relationshipspairslash.isin(keep_df.relationshipspairslash)]
top30_type = ds9.strtotype(top30_df, 'relationshipspairslash')
top30t_df = pd.DataFrame()
top30t_df['relationshipspairslash'] = top30_df['relationshipspairslash'].astype(top30_type)

# Graph
sns.set(rc = {'figure.figsize':(23,20)})
fig = sns.countplot(data=top30t_df, y='relationshipspairslash')
fig.set(title='Number of works in which a top-30 slash-pairing appeared')
fig.set(xlabel='Works')
fig.set(xticklabels=[])
fig.tick_params(bottom=False)
fig.set(ylabel='Top-30 slash-pair of characters')
fig.figure.savefig('frequencygraph-relationshipspairslash-top30.svg')
fig
```

```python
# Remove the most popular pair by name
top30t_df['relationshipspairslash'] = top30t_df[top30t_df.relationshipspairslash != 'Elim Garak/Julian Bashir']
# But do plot their position on the graph so it can be vertically compared
top30t_df['relationshipspairslash'] = top30t_df['relationshipspairslash'].astype(top30_type)

# Graph
sns.set(rc = {'figure.figsize':(23,20)})
fig = sns.countplot(data=top30t_df, y='relationshipspairslash')
fig.set(title='Number of works in which a top-30 slash-pairing appeared, without most frequent')
fig.set(xlabel='Works')
fig.set(xticklabels=[])
fig.tick_params(bottom=False)
fig.set(ylabel='Top-30 slash-pair of characters')
fig.figure.savefig('frequencygraph-relationshipspairslash-top30-minus1.svg')
fig
```
