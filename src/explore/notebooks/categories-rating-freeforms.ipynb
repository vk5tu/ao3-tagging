# Freeforms by frequency for (categories,ratings)



```python
import pandas as pd
import numpy as np

import ds9

df = ds9.df()
```

```python
# Explode column `collect`
def explode(df, column):
    explode_df = pd.DataFrame()
    for index, values in zip(df.index, df[column]):
        for i in values:
            row = df[df.index == index].copy()
            row[column] = i
            if 'id' not in row.columns:
                row['id'] = index
            explode_df = pd.concat([explode_df, row], ignore_index=True)
    return explode_df
```

```python
# Drop uninteresting columns
# This runs slowly, so instead itemise the column names, and then remove these from
# the list, then do a in_place drop of the columns.
drop_columns = df.columns.values.tolist()
drop_columns.remove('categories')
drop_columns.remove('rating')
drop_columns.remove('freeforms')
df.drop(drop_columns, axis=1, inplace=True)
```

```python
df = explode(df, 'categories')
```

```python
cats = list(df.categories.unique())
for c in cats:
    cat_df = df[df['categories'] == c]
    rats = cat_df['rating'].cat.categories
    for r in rats:
        print('Freeforms for {}, {}'.format(c, r))
        filename = ds9.csvfilename('categories-rating-freeforms', c, r)
        print(filename)
        print()
        freeform_df = cat_df[cat_df['rating'] == r]
        freeform_df = explode(freeform_df, 'freeforms')
        freq_df = freeform_df['freeforms'].value_counts()
        freq_df.to_csv(filename)
        print(freq_df.head(25))
        print()
```
